```python
#draw a bunch of cicles

pts = gpd.GeoSeries([Point(x, 0) for x in range(10)])
circles = pts.apply(lambda x: x.buffer(.5*np.random.rand()))

plt.figure(figsize=(6, 1))
ax1 = plt.subplot(1,1,1)
circles.plot(axes=ax1)
plt.axis('equal')
plt.box('off')
plt.yticks([])
plt.xticks([])

plt.savefig('Circles.png')
```

```python
#generate circles within a box
xmin, xmax, ymin, ymax = 0, 1, 0, 1

num_circles = 30
max_area = .005

xc = (xmax - xmin) * np.random.random(num_circles) + xmin
yc = (ymax - ymin) * np.random.random(num_circles) + ymin
pts = gpd.GeoSeries([Point(x, y) for x, y in zip(xc, yc)])

circles = pts.apply(lambda x: x.buffer(np.sqrt(max_area/np.pi * np.random.rand())))



plt.figure(figsize=(5, 5))
ax1 = plt.subplot(1,1,1)
circles.plot(axes=ax1)
plt.box('off')
plt.yticks([])
plt.xticks([])
plt.title('Vulnerabilities in the Code Base', fontsize=16)
plt.xlim(-.1, 1.1)
plt.ylim(-.1, 1.1)
ax1.add_patch(Rectangle((0,0), 1, 1, facecolor="grey", alpha=.25, zorder=0));

plt.savefig('Vuln_scatter_in_code.png')
```

```python
#draw cross-hairs for a 'search'
xmin, xmax, ymin, ymax = 0, 1, 0, 1

num_circles = 30
max_area = .005

xc = (xmax - xmin) * np.random.random(num_circles) + xmin
yc = (ymax - ymin) * np.random.random(num_circles) + ymin
pts = gpd.GeoSeries([Point(x, y) for x, y in zip(xc, yc)])

circles = pts.apply(lambda x: x.buffer(np.sqrt(max_area/np.pi * np.random.rand())))

x_seek = np.random.rand()
y_seek = np.random.rand()

plt.figure(figsize=(5, 5))
ax1 = plt.subplot(1,1,1)
circles.plot(axes=ax1)
plt.box('off')
plt.yticks([])
plt.xticks([])
plt.title('Vulnerabilities in the Code Base', fontsize=16)
plt.xlim(-.1, 1.1)
plt.ylim(-.1, 1.1)
ax1.add_patch(Rectangle((0,0), 1, 1, facecolor="grey", alpha=.25, zorder=0));

plt.hlines(y_seek, xmin, xmax)
plt.vlines(x_seek, ymin, ymax)

plt.savefig('Vuln_scatter_in_code_with_seek.png')
```
